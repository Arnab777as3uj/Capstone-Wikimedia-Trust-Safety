## Combining the annotated comments into one df

```python
df_c_1 = pd.read_csv("/home/ec2-user/SageMaker/bucket/wiki_trust/revisions_data/tsvs_google/aggression_annotated_comments.tsv",sep='\t')
df_c_1
```

```python
df_c_2 = pd.read_csv("/home/ec2-user/SageMaker/bucket/wiki_trust/revisions_data/tsvs_google/attack_annotated_comments.tsv",sep='\t')
df_c_2
```

```python
df_c_3 = pd.read_csv("/home/ec2-user/SageMaker/bucket/wiki_trust/revisions_data/tsvs_google/toxicity_annotated_comments.tsv",sep='\t')
df_c_3
```

```python
df_c_3 = df_c_3.astype({"rev_id": int})
df_c_3
```

```python
frames = [df_c_1,df_c_2,df_c_3]
df_c = pd.concat(frames)
df_c
```

```python
df_c = df_c[['rev_id','comment','ns']]
df_c
```

## Removing Duplicates

```python
df_c[df_c['rev_id'].isin(df_c[df_c.duplicated(subset='rev_id')==True]['rev_id'])].sort_values(by=['rev_id'])
```

```python
df_c[df_c.duplicated(subset='rev_id')==True].sort_values(by=['rev_id']).shape[0]
```

```python
df_c=df_c[df_c.duplicated(subset='rev_id')==False]
df_c.shape[0]
```

```python
df_c[df_c.duplicated(subset='rev_id')==True]
```

## Adding the annotation labels to the main df

```python
df_c = pd.merge(df_c,df1,how="left",left_on=["rev_id"],right_on=["rev_id"])
df_c.head(10)
```

```python
df_c = pd.merge(df_c,df2,how="left",left_on=["rev_id"],right_on=["rev_id"])
df_c.head(10)
```

```python
df_c = pd.merge(df_c,df3,how="left",left_on=["rev_id"],right_on=["rev_id"])
df_c.head(10)
```

```python
df_c.sample(10)
```

#### Removing null values

```python
df_c.isna().sum()
```

```python
df_c = df_c.fillna(0.0)
df_c.head(20)
```

```python
df_c.isna().sum()
```

#### Adding blocked label if any other of 3 label scores >= 0.5

```python
df_c['bl']= np.where((df_c['aggression']>=0.5)|(df_c['attack']>=0.5)|(df_c['toxicity']>=0.5),1,0)
df_c
```

```python
df_c['bl'].value_counts()
```

```python
df_c[df_c['bl']==1]
```

#### Removing NEWLINE_TOKEN

```python
df_c.comment = df_c.comment.str.strip()
```

```python
df_c.comment = df_c.comment.str.replace('NEWLINE_TOKEN',"")
```

```python
df_c[df_c['bl']==1]
```

### Formatting to fit the Text modelling corpus format

```python
del df_c['ns']
del df_c['aggression']
del df_c['attack']
del df_c['toxicity']
df_c
```

```python
df_c[df_c['bl']==1]
```

```python
df_c = df_c[['rev_id','bl','comment']]
df_c
```

```python
df_c=df_c.rename(index=str,columns={"rev_id": "username","comment":"text"})
df_c
```
